# DATA SPLITTING FOR MACHINE LEARNING

Here we break the data into smaller data frames x and y so that we can split them into training and testing data to create machine learning model

```python
x = heart_disease.drop(['condition'],axis=1)
```

```python
x
```

```python
y = heart_disease['condition']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=4)
```

# LOGISTIC REGRESSION AND MACHINE LEARNING MODEL

```python
from sklearn.linear_model import LogisticRegression

#confusion matrix is a 2d array that is used to compare predicted data to actual data. Instances include true positive, true negative,
#false positive, and false negative

#classification report measures the quality of predictions against actual data. It contains accuracy which is important
#for us. Accuracy is the measure of how many data points were predicted correctly from all classes (+ve or -ve)

from sklearn.metrics import classification_report , confusion_matrix
```

```python
logistic_model = LogisticRegression(random_state = 0)
logistic_model.fit(x_train, y_train)
```

```python
y_predict =logistic_model.predict(x_test)
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

the below heatmap shows that out of 90 data points, 45 were true negatives, 28 were true positives, 11 were false positives and 6 were false negatives 

```python
sns.heatmap(cm ,annot=True)
```

the below heatmap shows a confusion matrix in terms of percentages

```python
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

# IMPROVING THE MODEL

```python
logistic_model.fit(x_train,y_train)
```

```python
y_predict =logistic_model.predict(x_test)
```

```python
cn = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cn,  annot = True)
```

```python
print(classification_report(y_test,y_predict))
```

The accuracy we recieved is 81% or 81.1% for exact accuracy. The accuracy is (45+28)/(45+28+6+11) which is 0.811

# Using KNN classifier or K - nearest neighbours classification algorithm



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report , confusion_matrix
```

Since we are calculating standard euclidean distance we will use minkowski metric with a power parameter of 2

```python
knn_model = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski', p = 2)
knn_model.fit(x_train, y_train)
```

```python
y_predict =knn_model.predict(x_test)
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm ,annot=True)
```

```python
from sklearn import metrics
print("Accuracy for training set: ", metrics.accuracy_score(y_train, knn_model.predict(x_train)))
print("Accuracy for test set : ", metrics.accuracy_score(y_test, y_predict))
```

```python
print(classification_report(y_test,y_predict))
```

In order to get higher accuracy, you can change the number of neighbours and run the algorithm again
